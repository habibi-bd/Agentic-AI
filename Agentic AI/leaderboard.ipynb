```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Setup Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.kaggle.com/competitions/playground-series-s5e10/leaderboard")

# Wait for page to load JS content
time.sleep(5)  # increase if page is slow

# Find leaderboard rows
rows = driver.find_elements(By.CSS_SELECTOR, "ul.MuiList-root li.MuiListItem-root")

data = []
for row in rows[1:]:  # skip header
    cols = row.find_elements(By.CSS_SELECTOR, "span")
    if len(cols) >= 6:
        rank = cols[0].text
        team = cols[1].text
        members = cols[2].text
        score = cols[3].text
        entries = cols[4].text
        last = cols[5].text
        join = cols[6].text if len(cols) > 6 else ""
        data.append([rank, team, members, score, entries, last, join])

df = pd.DataFrame(data, columns=["Rank", "Team", "Members", "Score", "Entries", "Last", "Join"])
print(df.head())

driver.quit()
```

```python
df.to_csv("leaderboard.csv", index=False)
```
